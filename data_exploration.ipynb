```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
version = '0.1.3'
path = 'data/'
file_name = path + 'novels_' + version + '.csv'
```

```python
df = pd.read_csv(file_name)
df.head(2)
```

```python
df.shape
```

```python
df.info()
```

```python
# plot the number of novels per year
df['start_year'].value_counts().sort_index().plot(kind='bar', figsize=(15, 5))
plt.title('Number of novels per year')
plt.xlabel('Year')
plt.ylabel('Number of novels')
plt.show()
```

```python
# most used genres
genres = df['genres']
# each value is a array with strings
genres_list = ",".join(genres.tolist())
genres_list = genres_list.replace('[', '').replace(']', '').replace("'", '').replace(' ', '')
genres_list
```

```python
'''
make this in 1 line:

genres = df['genres']
genres_list = ",".join(genres.tolist())
genres_list = genres_list.replace('[', '').replace(']', '').replace("'", '').replace(' ', '').split(',')
'''
genres_list = ','.join(df['genres'].to_list()).replace('[', '').replace(']', '').replace("'", '').replace(' ', '').split(',')
genres_list
```

```python
genres_list = genres_list.split(',')
#plot
plt.figure(figsize=(15, 5))
sns.countplot(genres_list)
plt.title('Most used genres')
plt.xlabel('Genres')
plt.ylabel('Number of novels')
plt.xticks(rotation=90)
plt.show()
```

```python
# plot with bar chart
genres_list = pd.Series(genres_list)
genres_list.value_counts().plot(kind='bar', figsize=(15, 5))
plt.title('Most used genres')
plt.xlabel('Genres')
plt.ylabel('Number of novels')
plt.xticks(rotation=90)
plt.show()
```

```python
# plot with pie chart
genres_list.value_counts().plot(kind='pie', figsize=(15, 5))
plt.title('Most used genres')
plt.xlabel('Genres')
plt.ylabel('Number of novels')
plt.xticks(rotation=90)
plt.show()
```

```python
# get unique genres, authors, tags, original languages and number of chapters
genres = ','.join(df['genres'].to_list()).replace('[', '').replace(']', '').replace("'", '').replace(' ', '').split(',')
genres = pd.Series(genres).unique()
authors = df['authors'].unique()
tags = ','.join(df['tags'].to_list()).replace('[', '').replace(']', '').replace("'", '').replace(' ', '').split(',')
tags = pd.Series(tags).unique()
original_languages = df['original_language'].unique()

print('Number of genres: ', len(genres))
print('Number of authors: ', len(authors))
print('Number of tags: ', len(tags))
print('Number of original languages: ', len(original_languages))
```
